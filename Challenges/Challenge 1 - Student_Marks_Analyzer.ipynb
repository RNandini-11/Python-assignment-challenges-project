'''
Challenge: Student Marks Analyzer
Create a Python program that allows a user to input student names along with their marks
and then calculates useful statistics.
Your program should:
1. Let the user input multiple students with their marks (name + integer score).
2. After input is complete, display:
   - Average marks
   - Highest marks and student(s) who scored it
   - Lowest marks and student(s) who scored it
   - Total number of students

Bonus:
- Allow the user to enter all data first, then view the report
- Format output clearly in a report-style layout
- Prevent duplicate student names
'''

```python
class StudentMarksAnalyzer:
    def __init__(self):
        self.stud=[]  #list if dict

    def add_stud(self):
        self.no=int(input('How many student to enter details? '))
        for i in range(1,self.no+1):
            print(f"\n---Enter details of student {i}---")
            name = input('Enter student name: ')
            while any(s['Name'].lower() == name.lower() for s in self.stud):
                print(f"Error: Student {name} already exists!")
                name = input('Enter name again: ')
            m1 = int(input('Enter marks of Sub1: '))
            m2 = int(input('Enter marks of Sub2: '))
            m3 = int(input('Enter marks of Sub3: '))
            m4 = int(input('Enter marks of Sub4: '))
            m5 = int(input('Enter marks of Sub5: '))
            marks=[m1,m2,m3,m4,m5]
            stud_data = {'Name': name, 'Marks': marks}
            self.stud.append(stud_data)
            print(f'___{name} student added successfully!\n')

    def cal_marks(self):
      print("\n----- Student Report -----\n")
      all_totals = []
      for student in self.stud:
          total = sum(student['Marks'])
          avg = total / len(student['Marks'])
          all_totals.append(total)
          print(f"Name: {student['Name']} | Total: {total} | Average: {avg:.2f}")
      highest = max(all_totals)
      lowest = min(all_totals)

      top_students = []
      for s in self.stud:
          total = sum(s['Marks'])
          if total == highest:
              top_students.append(s['Name'])

      low_students = []
      for s in self.stud:
          total = sum(s['Marks'])
          if total == lowest:
              low_students.append(s['Name'])

      overall_avg = sum(all_totals) / len(all_totals)
      print("\n----- Summary -----")
      print(f"Overall Average Marks: {overall_avg:.2f}")
      print(f"Highest Marks: {highest} | Students: {', '.join(top_students)}")
      print(f"Lowest Marks: {lowest} | Students: {', '.join(low_students)}")
      print(f"Total Students: {len(self.stud)}")

    def display(self):
        print("\n========== STUDENT MARKS REPORT ==========\n")
        for i, student in enumerate(self.stud, start=1):
            print(f"Student {i}: {student['Name']}")
            print("-----------------------------------")
            for idx, mark in enumerate(student['Marks'], start=1):
                print(f"Subject {idx}: {mark}")
            total = sum(student['Marks'])
            avg = total / len(student['Marks'])
            print(f"Total Marks : {total}")
            print(f"Average     : {avg:.2f}")
            print("-----------------------------------\n")

        print("=========================================")
        self.cal_marks()

s=StudentMarksAnalyzer()
s.add_stud()
s.display()
```
