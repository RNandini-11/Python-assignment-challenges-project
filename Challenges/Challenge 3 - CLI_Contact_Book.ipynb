Challenge: CLI Contact Book (CSV-Powered)

Create a terminal-based contact book tool that stores and manages contacts using a CSV file.

Your program should:
1. Ask the user to choose one of the following options:
   - Add a new contact
   - View all contacts
   - Search for a contact by name
   - Exit
2. Store contacts in a file called `contacts.csv` with columns:
   - Name
   - Phone
   - Email
3. If the file doesn't exist, create it automatically.
4. Keep the interface clean and clear.

Example:
Add Contact
View All Contacts
Search Contact
Exit

Bonus:
- Format the contact list in a table-like view
- Allow partial match search
- Prevent duplicate names from being added

```python
import csv
from tabulate import tabulate
import re
import os

class Contact:
  def __init__(self):
    try:
      if not os.path.exists("contacts.csv"):
          with open("contacts.csv","w", newline='') as f:
              writer = csv.writer(f)
              writer.writerow(["Name", "Phone", "Email"])
      print('Contacts added successfully...!')
    except Exception as e:
      print('Error: ',e)

  def menu(self):
    while True:
      print("\n====== Menu for Contact  ======")
      print('\n1.Add a new contact')
      print('2.View all contacts')
      print('3.Search for a contact by name')
      print('4.Exit')
      try:
          choice = int(input('Enter your choice: '))
      except ValueError:
          print("Invalid input!")
          continue

    if choice==1:
      self.add_contact()
    elif choice==2:
      self.view_contacts()
    elif choice==3:
      self.search_contact()
    elif choice==4:
      exit()
    else:
      print('invalid choice...!')


  def add_contact(self):
    #for valid no of input contact
    while True:
      try:
          self.no=int(input('How many number of contacts you want to add? '))
          if self.no <= 0:
              print("Enter a positive number...!")
              continue
          break
      except Exception as e:
          print('Error: ',e)

    for i in range(self.no):
      print('\n---Enter contact '+str(i+1),'---')

      #for valid name
      while True:
        name=input('Enter name: ')
        if not name:
            print("Name cannot be empty...!")
            continue
        if any(char.isdigit() for char in name):
            print("Name cannot contain numbers...!")
            continue
        with open("contacts.csv", "r") as f:
            if any(row['Name'].lower() == name.lower() for row in csv.DictReader(f)):
                print(f"Student {name} already exists in the file!")
                continue
        break

      #for valid phone no
      while True:
        phone=input('Enter phone: ')
        if not phone:
            print("Phone cannot be empty...!")
            continue
        if not phone.isdigit():
            print("Phone number should contain only digits...!")
            continue
        if len(phone) != 10:
            print("Phone number should be exactly 10 digits...!")
            continue
        break

      #for valid email
      while True:
        email=input('Enter email: ')
        if not email:
            print("Email cannot be empty...!")
            continue
        if not re.match(r'^\S+@\S+\.\S+$', email):
            print("Invalid email format...!")
            continue
        break

      try:
          with open("contacts.csv","a", newline='') as f:
              writer = csv.writer(f)
              writer.writerow([name, phone, email])
          print('Contact added successfully...!')
      except Exception as e:
        print('Error: ', e)
    self.menu()

  def view_contacts(self):
    print("\n============ Contact Details ============\n")
    try:
      with open('contacts.csv', mode='r') as file:
        reader = csv.reader(file)
        data = list(reader)
        print(tabulate(data[1:], headers=data[0], tablefmt="grid"))
        print('Contacts viewed successfully...!')
    except Exception as e:
      print('Error: ',e)
    self.menu()

  def search_contact(self):
    search_name=input('Enter name: ').lower()
    try:
      with open("contacts.csv","r") as f:
        reader = csv.reader(f)
        data = list(reader)
        if not data:
          print("The file is empty.")
        else:
          header = data[0]
          rows = data[1:]
          filtered_results = [row for row in rows if search_name in row[0].lower()]
          if filtered_results:
              print(tabulate(filtered_results, headers=header, tablefmt="grid"))
              print('Contacts searched successfully...!')
          else:
              print(f"No results found for '{search_name}'.")
    except Exception as e:
      print('Error: ',e)
    self.menu()

  def exit(self):
    print('Thank you for using our program')
    exit()

c=Contact()
```
