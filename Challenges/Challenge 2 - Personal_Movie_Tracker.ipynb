Challenge:  Personal Movie Tracker with JSON

Create a Python CLI tool that lets users maintain their own personal movie database, like a mini IMDb.

Your program should:
1. Store all movie data in a `movies.json` file.
2. Each movie should have:
   - Title
   - Genre
   - Rating (out of 10)
3. Allow the user to:
   - Add a movie
   - View all movies
   - Search movies by title or genre
   - Exit the app

Bonus:
- Prevent duplicate titles from being added
- Format output in a clean table
- Use JSON for reading/writing structured data

```python
import json
import os
from tabulate import tabulate


class Movie:
    def __init__(self):
        if not os.path.exists("movies.json"):
            with open("movies.json", "w") as file:
                json.dump([], file)
        self.menu()

    def menu(self):
        while True:
            print("\n______ Menu for Movie ______")
            print("1. Add a movie")
            print("2. View all movies")
            print("3. Search movies by title or genre")
            print("4. Exit the app")

            try:
                choice = int(input('Enter your choice: '))
            except ValueError:
                print("Invalid input!")
                continue

            if choice == 1:
                self.add_movie()
            elif choice == 2:
                self.view_movies()
            elif choice == 3:
                self.search_movies()
            elif choice == 4:
                self.exit()
            else:
                print('invalid choice...!')
            break

    def add_movie(self):
        # Load existing data
        with open("movies.json", "r") as file:
            movie_list = json.load(file)

        # for valid no of input movie
        while True:
            try:
                self.no = int(input('How many number of movies you want to add? '))
                if self.no <= 0:
                    print("Enter a positive number...!")
                    continue
                break
            except ValueError:
                print('Invalid input! Please enter a number.')

        for i in range(self.no):
            print('\n---Enter Movie ' + str(i + 1), '---')

            # title validation
            while True:
                title = input("Enter movie title: ")
                if not title:
                    print("Title cannot be empty...!")
                    continue
                if title.isdigit():
                    print("Title cannot contain numbers...!")
                    continue
                if any(m['Title'].lower() == title.lower() for m in movie_list):
                    print(f"Movie {title} already exists in the file!")
                    continue
                break

            # genre validation
            while True:
                genre = input("Enter genre: ")
                if not genre:
                    print("Genre cannot be empty...!")
                    continue
                if any(char.isdigit() for char in genre):
                    print("Genre cannot contain numbers...!")
                    continue
                break

            # rating validation
            while True:
                try:
                    rating = float(input("Enter rating(0-10): "))
                    if rating <= 0 or rating > 10:
                        print("Enter a positive number...!")
                        continue
                    break
                except Exception as e:
                    print('Error: ', e)

            #add movie
            movie_list.append({
                "Title": title,
                "Genre": genre,
                "Rating (out of 10)": rating,
            })

        # Save updated data
        try:
            with open("movies.json", "w") as file:
                json.dump(movie_list, file, indent=4)
        except Exception as e:
            print('Error: ', e)
        print('Movie added successfully...!')
        self.menu()

    def view_movies(self):
        try:
            with open("movies.json", "r") as f:
                movies = json.load(f)
                if not movies:
                    print("No movies available.")
                else:
                    print("\n============ All Movies ============\n")
                    table = [[m['Title'], m['Genre'], m['Rating (out of 10)']] for m in movies]
                    print(tabulate(table, headers=["Title", "Genre", "Rating (out of 10)"], tablefmt="grid"))
                    print('Movies viewed successfully...!')
        except Exception as e:
            print('Error: ', e)
        self.menu()

    def search_movies(self):
        try:
            with open("movies.json", "r") as f:
                movies = json.load(f)
        except Exception as e:
            print('Error: ', e)

        # Valid search input
        if not movies:
            print("No movies available.")
        else:
            while True:
                search = input("\nEnter movie title or genre: ").lower()
                if not search:
                    print("Search cannot be empty!")
                    continue
                break

            results = []
            for m in movies:
                if (search in m["Title"].lower()) or (search in m["Genre"].lower()):
                    results.append([m["Title"], m["Genre"], m["Rating (out of 10)"]])
            print(tabulate(results, headers=["Title", "Genre", "Rating"], tablefmt="grid"))
            print('Contacts searched successfully...!')
            self.menu()

    def exit(self):
        print("Thank you for using Movie Tracker!")
        exit()


m = Movie()
```
